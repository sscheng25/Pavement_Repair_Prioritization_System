```python
# Get all amenities in El Paso
amenities = ox.geometries_from_place("El Paso, TX", tags={"amenity": True})
len(amenities)
```

```python
amenities.head()
```

```python
amenities['amenity'].unique()
```

```python
count = amenities[['amenity', 'name']].groupby('amenity').count().reset_index().sort_values(by = 'name', ascending = False)
# count.head(50)
```

```python
food_drink = ['restaurant', 'fast_food', 'cafe', 'bar', 'ice_cream', 'pub', ]
entertainment = ['arts_centre', 'cinema', 'theatre', ]
car_facility = ['fuel', 'car_rental', 'car_wash', 'parking', 'parking_space', ]
```

```python
all_amenity = food_drink + entertainment + car_facility
```

```python
# Get bar, pub, and restaurant features in Philadelphia
food_and_drink = ox.geometries_from_place("El Paso, TX", tags={"amenity": food_drink})

len(food_and_drink)
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

ax = food_and_drink.plot(ax=ax)
ax.set_axis_off()
```

```python
food_and_drink
```

```python
all_df = ox.geometries_from_place("El Paso, TX", tags={"amenity": all_amenity})
all_df = all_df.reset_index()
```

```python
all_df = all_df[['element_type', 'osmid', 'amenity', 'gnis:county_name', 'name', 'geometry']]
```

```python
all_df.dtypes
```

```python
all_df.geometry = all_df.geometry.centroid
```

```python
all_df['lon'] = all_df.geometry.apply(lambda p: p.x)
all_df['lat'] = all_df.geometry.apply(lambda p: p.y)
```

```python
save_df = all_df[['element_type', 'osmid', 'amenity', 'gnis:county_name', 'name', 'lon', 'lat']]
```

```python
save_df.to_csv('OSM_amenity.csv', index=False)
```
